## Univariate (simple) Regression 

```python
X = data['Size']
Y = data['Price']
```

```python
X
```

```python
Y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X,Y)
plt.axis([0, 1500, 0, 1500000])
plt.xlabel('House Size (sq.ft)')
plt.ylabel('House Price')
plt.show()
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()
```

```python
reg.summary()
```

# Multivariate Regression

```python
mult_data = pd.read_excel('C:/Nijat/Nicat/Python/[FreeTutorials.Us] python-for-finance-investment-fundamentals-data-analytics/16-part-ii-finance-multivariate-regression-analysis/Housing.xlsx')
```

```python
mult_data
```

```python
X = mult_data[['House Size (sq.ft.)', 'Number of Rooms', 'Year of Construction']]
Y = mult_data['House Price']
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()

reg.summary()
```
